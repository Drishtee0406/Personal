14.1 Satellite Radio:

Association rule mining cannot be used on the Satellite dataset due to inherent structural and analytical limitations:

1️. Data format mismatch – Association rules are ideal for transactional data, where a record represents a collection of items that usually occur together (e.g., shopping carts). The Satellite dataset, however, includes structured attributes like zip conversion, homeownership, income, and wealth, that record individuals instead of co-occurring items.

2️. Lack of Alignment with the Purpose of Analysis – This dataset is likely meant for customer demographic analysis, predicting behavior, or clustering like-minded people rather than identifying patterns of items co-occurring. Classification, regression, or clustering would be more appropriate approaches to these tasks.

3️. Structural Incompatibility – The data consists of numerical and categorical attributes, as compared to a transactional format with discrete co-occurring items. Since there is no market basket-type data where items occur or do not occur in transactions, association rule mining is not suitable for this data.

```python
data = {
    'Intro': [1, 2, 3, 4, 5]
    ,'Data': ['DOE, SW, DataMining', 'DOE, SW', 'Forecast, Regression', 'DataMining, Survey, CatData', 'CatData, Forecast, SW']
}

df = pd.DataFrame(data)
df['Data'] = df['Data'].apply(lambda x: x.split(',')) #helps in purchasing item patterns
```

```python

all_word_rows = df['Data'].tolist() 
all_words = set() # Initialize an empty set to store unique words (items)

for row in all_word_rows:
    for word in row:
        all_words.add(word)
```

```python
def generate_itemsets(all_words):
    all_itemsets = set()

    for x in range(1, len(all_words) + 1):
        all_itemsets.update(combinations(all_words, x)) # creates number of combiantions
        
    return all_itemsets
```

```python
def generate_rules(item_set):
    rules = [] # List to store the generated rules
    itemset_length = len(item_set)
    set_itemset = set(item_set)

    for n in range(1, itemset_length): # Convert item_set into a set for set operations
        for combo in combinations(set_itemset, n):
            antecedent = set(combo)
            consequent = set_itemset - antecedent
            rules.append((antecedent, consequent, set_itemset))
             # generate rules for all possible subsets of item_set 
    return rules
```

```python
binary_df = pd.DataFrame(columns=list(all_words))# Create an empty DataFrame with column names
```

```python
for idx, row in enumerate(all_word_rows):
    for word in row:  # Iterate over each word in the current row
        binary_df.loc[idx, word] = 1
```

```python
binary_df = binary_df.map(lambda x : True if x == 1 else False) # Convert all values in binary_df to boolean values (True/False)
```

```python
all_itemsets = generate_itemsets(all_words) # Generate all possible itemsets from the given set of words
```

```python
all_rules = [] # Initialize an empty list to store all generated rules

for itemset in all_itemsets:
    all_rules.extend(generate_rules(itemset)) # and extend the all_rules list by adding the generated rules
```

```python
support_dict = {}  # Dictionary to store the support count of each itemset

for row in all_itemsets:
    total_cols = len(row)

    for idx in range(len(binary_df)):
        if (binary_df.loc[idx, list(row)].sum() == total_cols):
            support_dict[row] = support_dict.get(row, 0) + 1
```

```python
support_list = [(set(k), v / len(binary_df)) for k, v in support_dict.items()]
```

```python
rules_df = pd.DataFrame(all_rules, columns=['antecedent', 'consequent', 'itemset'])
```

```python
def get_support_value(itemset):
    for support_metric in support_list:
        if support_metric[0] == itemset:
            return support_metric[1]
    return 0
```

```python
rules_df['antecedent_support'] = rules_df['antecedent'].apply(get_support_value)
rules_df['consequent_support'] = rules_df['consequent'].apply(get_support_value)
rules_df['itemset_support'] = rules_df['itemset'].apply(get_support_value)
rules_df['confidence'] = rules_df['itemset_support'] / rules_df['antecedent_support']
rules_df['lift'] = rules_df['confidence'] / rules_df['consequent_support']
```

```python
minimum_support = .1

rules_df = rules_df[rules_df['itemset_support'] > minimum_support]
```

The analysis shows strong correlation between "CatData" and "DataMining", indicating that users interested in "CatData" almost always visit "DataMining", highlighting an opportunity to bundle these topics together, improve recommendations, and increase user engagement further.
